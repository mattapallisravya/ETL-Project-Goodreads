```python
# Imports
import pandas as pd
```

```python
# Read the CSV file and put it into a dataframe

csv_file = 'Resources/original_book_data.csv'
book_df = pd.read_csv(csv_file)
book_df.head()
```

Notes: Price is funky and has some blanks

```python
# Store the columns we want in a new DF
book_df2 = book_df[['title', 'author','rating', 'language', 'isbn',
    'genres', 'pages', 'publisher', 'numRatings', 'ratingsByStars', 'price','awards']].copy()

# Filter DF to only include books in English
book_df2 = book_df2[book_df2['language'] == 'English']

book_df2.head()
```

```python
# Determine the column dtypes
book_df2.dtypes
```

```python
## convert the Title, Author, Language column to a string
book_df2['title'] = book_df2['title'].astype('string')
book_df2['author'] = book_df2['author'].astype('string')
book_df2['language'] = book_df2['language'].astype('string')


# first convert pages to a string
book_df2['pages'] = book_df2['pages'].astype('string')
## then remove '1 page' values
book_df2.drop(book_df2[book_df2['pages'] == '1 page'].index, inplace = True)
## convert the pages to an integer
book_df2['pages'] = book_df2['pages'].astype('int64')
#book_df2['pages'].unique()
```

```python
## convert the ISBN column to a string
book_df2['isbn'] = book_df2['isbn'].astype('string')

# drop books with the ISBN of 9999999999999
book_df2.drop(book_df2[book_df2['isbn'] == '9999999999999'].index, inplace = True)

# remove duplicate ISBN numbers
book_df2.drop_duplicates(subset=['isbn'], keep = 'first')

book_df2.head()
```

```python

# export to csv to share with the team

book_df2.to_csv('clean_book_data_1.csv', index=False)
```
